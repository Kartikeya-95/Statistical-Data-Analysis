```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
pop = pd.read_csv('../../DataSets/Population/Total_Population/API_SP.POP.TOTL_DS2_en_csv_v2_63973.csv', skiprows = 4)
pop_male = pd.read_csv('../../DataSets/Population/Total_Male/API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_42292.csv', skiprows = 4)
pop_female = pd.read_csv('../../DataSets/Population/Total_Female/API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_41558.csv', skiprows = 4)
```

```python
# Importing table having list of all countries along with ISO codes
countries = pd.read_html('https://countrycode.org/')
```

```python
# A list of dataframe is returned by read_html(), so we take first dataframe 
countries = countries[0]
countries.head()
```

```python
countries.shape
```

```python
# Filtering 'ISO CODES' column to keep only 3 letter code

countries['ISO CODES'] = countries['ISO CODES'].apply(lambda x : x[-3:])
countries['ISO CODES']
```

```python
pop.head()
```

```python
pop_male.head()
```

```python
pop_female.head()
```

```python
pop.shape, pop_male.shape, pop_female.shape
```

```python
# Dropping 'Indicator Name', 'Indicator Code' and 'Unnamed:63' columns from dataframes (2nd, 3rd and last)

pop.drop(pop.columns[[2,3,-1]], axis = 1, inplace = True)
pop_male.drop(pop_male.columns[[2,3,-1]], axis = 1, inplace = True)
pop_female.drop(pop_female.columns[[2,3,-1]], axis = 1, inplace = True)
```

```python
# Checking columns
pop.columns, pop_male.columns, pop_female.columns
```

```python
# checking datatypes of columns
pop.dtypes.values, pop_male.dtypes.values, pop_female.dtypes.values
```

```python
# Keeping only those rows where country code matches from 'ISO CODES' in 'countries' table. This filtering is required as 
# there are other entries in 'Country Name' columns like Low Income, World etc. We intend to keep only countries.

pop = pop[pop['Country Code'].isin(countries['ISO CODES'])]
pop_male = pop_male[pop_male['Country Code'].isin(countries['ISO CODES'])]
pop_female = pop_female[pop_female['Country Code'].isin(countries['ISO CODES'])]
```

```python
# Checking shape again 
pop.shape, pop_male.shape, pop_female.shape
```

```python
# Checking index locations where null values are present, (row-wise)
pop_null = pop.isnull().sum(axis = 1)
pop_male_null = pop_male.isnull().sum(axis = 1)
pop_female_null = pop_female.isnull().sum(axis = 1)

print('Total_Population : \n{}'.format(pop_null[pop_null > 0]))
print('Total_Male : \n{}'.format(pop_male_null[pop_male_null > 0]))
print('Total_Female : \n{}'.format(pop_female_null[pop_female_null > 0]))
```

```python
# Dropping rows where number of null values >= 30, we passed thresh = 36 (means rows having atlesat 36 non-NA values are kept)

pop.dropna(axis = 0, thresh = 36, inplace = True)
pop_male.dropna(axis = 0, thresh = 36, inplace = True)
pop_female.dropna(axis = 0, thresh = 36, inplace = True)
```

```python
# Checking null values once again
pop_null = pop.isnull().sum(axis = 1)
pop_male_null = pop_male.isnull().sum(axis = 1)
pop_female_null = pop_female.isnull().sum(axis = 1)

print('Total_Population : \n{}'.format(pop_null[pop_null > 0]))
print('Total_Male : \n{}'.format(pop_male_null[pop_male_null > 0]))
print('Total_Female : \n{}'.format(pop_female_null[pop_female_null > 0]))
```

```python
# Exporting cleaned dataframes to csv
pop.to_csv('../../DataSets/Population/Total_Population/Cleaned_Total_Population.csv', index = False)
pop_male.to_csv('../../DataSets/Population/Total_Male/Cleaned_Male_Population.csv', index = False)
pop_female.to_csv('../../DataSets/Population/Total_Female/Cleaned_Female_Population.csv', index = False)
```
